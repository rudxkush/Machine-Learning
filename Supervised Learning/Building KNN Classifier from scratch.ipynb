```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('/Users/rudxkush/Desktop/Daily Use/Kaggle Dataset/Social_Network_Ads.csv')
df.head()
```

```python
X = df.iloc[:, 2:4].values  # X : {Age, Est._salary}
y = df.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# Age and Estimated Salary have very different ranges, so feature scaling is required
# to prevent Salary from dominating the distance calculation
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
import operator
from collections import Counter

class KNearestNeighbour:
    def __init__(self, k):
        self.k = k

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        print("Training Done!")

    def predict(self, X_test):
        predictions = []
        for x in X_test:
            distance = {}
            counter = 0

            for i in self.X_train:
                distance[counter] = ((x[0] - i[0])**2 + (x[1] - i[1])**2) ** 0.5
                counter += 1

            distance = sorted(distance.items(), key=operator.itemgetter(1))
            label = self.classify(distance[:self.k])
            predictions.append(label)

        return predictions

    def classify(self, dist):
        label = []
        for i in dist:
            label.append(self.y_train[i[0]])

        return Counter(label).most_common()[0][0]
```

```python
classifier = KNearestNeighbour(k=3)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
