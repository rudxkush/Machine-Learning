```python
from matplotlib.pyplot import step
# To create synthetic dataset, we are going to use make_classification from sklearn
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41, hypercube= False,
                           class_sep=10)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.1
    for i in range(1000):
        j = np.random.randint(0, 100) # pick a random index from dataset
        y_hat = step(np.dot(X[j], weights)) # predict it's y value
        weights = weights + learning_rate *  (y[j] - y_hat) * X[j] # update weights

    return weights[0], weights[1:]
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
intercept, coef = perceptron(X, y)
```

```python
print(coef)
print(intercept)
```

```python
m = -(coef[0]/coef[1])
c = -(intercept/coef[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
