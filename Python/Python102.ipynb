```python
def print_item_details(*Args, **kwargs):
  print(Args)
  print(kwargs)

print_item_details(22, "blah", name = "itcy", weight = 23.34)
```

```python
## map
def even_or_odd(num):
  if(num%2 == 0):
    print("the number {} is even".format(num));
  else:
    print("the number {} is odd".format(num));

arr = [1, 2, 3, 5, 6, 7, 11, 10, 8]
list(map(even_or_odd, arr)) # Note : map() collects the return values of the function it applies
# in this case it None.
```

```python
## lambda
addition = lambda a, b : a + b
print(addition(12, 12))
```

```python
## filter function
arr = [1, 2, 3, 5, 6, 7, 11, 10, 8]
print(list(filter(lambda num : num%2 == 0, arr)))

def lst_square(lst):
    for i in range(len(lst)):
        lst[i] = lst[i] * lst[i]
    print(lst)

lst = [1, 2, 3, 4, 5, 6, 7]
lst_square(lst)
```

```python
## list comprehension
print([i*i for i in lst if i%2 == 0])
```
