## OOPs

```python
class Car:
  def __init__(self, window = None, door = None, enginetype = None):
    self.windows = window
    self.doors = door
    self.enginetype = enginetype

  def show_details(self):
    if self.windows is not None:
      print("windows:", self.windows)
    else:
      print("Not yet confirmed!")
    if self.doors is not None:
      print("doors:", self.doors)
    else:
      print("Not yet confirmed!")
    if self.enginetype is not None:
      print("enginetype:", self.enginetype)
    else:
      print("Not yet confirmed!")

  def vehicle_run_state(self):
    print("This variant runs on {}.".format(self.enginetype))
```

```python
car1 = Car()
```

```python
car1
```

```python
car1.windows = 6
car1.doors = 4
```

```python
print(car1.windows)
```

```python
car1.show_details()
```

```python
car1.enginetype = 'petrol'
```

```python
car1.vehicle_run_state()
```
