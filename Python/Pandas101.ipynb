```python
import pandas as pd
# It provides high performance, easy to use data structure and data analysis tools
import numpy as np
```

```python
df = pd.DataFrame(np.arange(0, 20).reshape(5, 4),
                  index= ["Row1", "Row2", "Row3", "Row4", "Row5"], columns=["Column1", "Column2", "Column3", "Column4"]);
# DataFrame takes: data + row names (index) + column names
# print(df);
df.iloc[0:6,0:5]
# Note : DataFrame is excel sheet with rows and columns whereas, Series is Single Strip of Data
```

```python
df.to_csv("Test1.csv"); # creates csv file
```

```python
# To access the elements
df.loc['Row1']
# df.loc['Row1', 'Column2']
# .loc : label based indexing (rows and columns by name)
# .iloc : position based indexing (rows and columns by number)
```

```python
df.iloc[3, 3]
```

```python
df.iloc[0:3, 0:2]
```

```python
df.iloc[:, 1:] # Exclude Column1
```

```python
df.iloc[:, 1:].values # convert dataframe into array
```

```python
df['Column1'].value_counts() # returns unique elements in that column along with their frequency
```

```python
df['Column1'].unique() # returns an array of all unique elements in that column
```

```python
# Quoting and Escape Characters.
data = 'a,b\n"hello, \\"Bob\\", nice to meet you", 5'
```

```python
from io import StringIO
pd.read_csv(StringIO(data), escapechar='\\', index_col = 0)
# StringIO treats a string like a file.
# If you see a backslash \\, treat the next character as literal text.
```

```python
import requests

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent": "Mozilla/5.0" # Add headers to mimic a real browser (some sites block default requests)
}

response = requests.get(url, headers=headers) # Send GET request to fetch page content
# Extract all HTML tables that contain the word "Country"
# header=0 means the first row is treated as column names
dfs = pd.read_html(response.text, match="Country", header=0, index_col=0)

# Number of tables found
len(dfs)
```
