```python
import numpy as np
# It provides high performance multi-dimensional array objects (ndarray)

# Examples:
# - 2D array (array of arrays of float) → weight matrix in attention mechanism
# - 3D array → batch_size x sequence_length x embedding_dimension tensor
```

```python
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]

arr = np.array(lst);
print(type(arr));
print(arr);
print(arr.shape)
```

```python
lst1 = [1, 2, 3, 4]
lst2 = [5, 6, 7, 8]
lst3 = [9, 10, 11, 12]

mat = np.array([lst1, lst2, lst3]);
print(type(mat));
print(mat);
print(mat.shape)
```

```python
print(arr.reshape(3, 3))
```

```python
print(mat)
print(mat[0:2, 0:2]) # matrix(start_row, end_row + 1, start_col, end_col + 1)
print(mat[2:, 2:])
```

```python
original = arr;
arr = np.arange(0,10,step=2);
print(arr)
print(original);
```

```python
np.linspace(1, 10, 50)
# Creates 50 equally spaced numbers between 1 and 10 (inclusive).
# Used in - Creating synthetic datasets, Plotting functions
```

```python
original[3:]=100
print(original)
```

```python
copy_arr = original
copy_arr[3:] = 500
print(copy_arr)
print(original) # As array is a reference type, both are accessing the same shared memory space
```

```python
copy_arr = original.copy() # Another memory block is allocated with the same values and shape as original array
copy_arr[3:] = 1000
print(copy_arr)
print(original)
```

```python
print(arr[arr > 1])
print(arr * 3)
```

```python
print(np.arange(0, 10).reshape(2, 5)) # create arrays and reshape
```
