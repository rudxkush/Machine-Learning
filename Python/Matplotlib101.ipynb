Matplotlib is a Python library used to create graphs and charts.</br>
It works well with NumPy and lets you embed plots into apps (Tkinter, Qt, etc.)</br>
- Easy for basic plots
- Add custom labels and text easily
- Full control over every part of the figure
- High-quality output (PNG, PDF, etc.)
- Highly customizable

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

plotting sample examples

```python
x = np.arange(0, 10);
y = np.arange(11, 21);
```

```python
a = np.arange(40, 50);
b = np.arange(51, 61);
```

```python
## plt scatter
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
# plt.savefig('graph.png')
```

```python
y = x*x
```

```python
plt.plot(x, y,'r*--', linestyle='dashed', linewidth=2, markersize=12)
plt.title('y = x*x is a parabola')
```

```python
# creating subplots
plt.subplot(2,2,1) # 2 rows, 2 columns, place it at 1st!
plt.plot(x,y,'r--',c='red')
plt.subplot(2,2,3)
plt.plot(x,y,'r--',c='green')
plt.subplot(2,2,4)
plt.plot(x,y,'r--',c='blue')
```

```python
np.pi
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('Sine Wave Form')
plt.plot(x,y)
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y_sine = np.sin(x);
y_cosine = np.cos(x);
plt.plot(x,y_sine)
plt.plot(x,y_cosine)
```

```python
# Create a 2-row, 1-column grid and select the first subplot
plt.subplot(2, 1, 1)
plt.plot(x, y_sine)
plt.title('Sine')


# Select the second subplot in the same 2x1 grid
plt.subplot(2, 1, 2)
plt.plot(x, y_cosine)
plt.title('Cosine')

plt.tight_layout() # plt.tight_layout() automatically adjusts spacing between the subplots
```

```python
# Bar plot
x1 = [2, 8, 10]
y1 = [11, 16, 9]
x2 = [3, 9, 11]
y2 = [6, 15, 7]
plt.bar(x1, y1, align='center')
plt.bar(x2, y2, color='green')
plt.title('Bar Chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Histogram
arr = np.array([22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
plt.hist(arr)
plt.title('Histogram')
```

```python
# Box Plot using matplotlib
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# Rectangular Box Plot
plt.boxplot(data, vert=True, patch_artist=True)
# small circles in the graph represent the outliers
plt.title('Box Plot')
```

```python
plt.boxplot(data, vert=False, patch_artist=False)
```

```python
# pie chart
labels = 'Sampreet', 'Aviroopa Dey', 'Aarushi', 'Diya'
data = 100, 200, 450, 250
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.3, 0.2) # explode 3rd slice
plt.pie(data, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.title('Pie Chart')
```
